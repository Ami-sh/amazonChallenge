```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

# Read the CSV file into a pandas DataFrame

df = pd.read_csv(r'Trainn.csv')
dfTest=pd.read_csv(r'testt.csv')
test_df=dfTest.copy()

# df=df.iloc[:,[4,5]]
# dfTest=dfTest.iloc[:,[1,2]]

X_train=df.iloc[:,[1,4]]
y_train=df.iloc[:,[2]]
X_test=dfTest.iloc[:,[2,4]]

def saveFile(y_pred):
    submission_df=test_df.copy()
    submission_df['PRODUCT_LENGTH']=y_pred
    submission_df=submission_df[['PRODUCT_ID','PRODUCT_LENGTH']]
    submission_df.set_index('PRODUCT_ID',inplace=True)
    submission_df.to_csv('sample_submission.csv')
# X_train
X_test
```

```python
# Create a KNN regression object
knn = KNeighborsRegressor(n_neighbors=3)

# Train the regression on the training data
knn.fit(X_train, y_train)

# Predict the values of the test data
y_pred = knn.predict(X_test)
saveFile(y_pred)
```
